```python
import matplotlib.pyplot as plt


demo = data.groupby("year")["amount"].sum()

plt.figure(figsize=(8,5))
plt.plot(demo.index, demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total Amount")
plt.title("Year Wise Amount Investment", color="blue", fontsize=15)
plt.xlabel("Year", color="green", fontsize=12)
plt.ylabel("Total Amount", fontsize=12)
plt.grid(linestyle=":", color="skyblue", linewidth=1.5)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data["city"].value_counts().head(10)

demo = data["city"].value_counts().head(10)
color =["blue","black","red","yellow","pink","green","purple","orange","brown","skyblue"]

plt.figure(figsize=(10,6))
plt.bar(demo.index, demo.values,color=color,alpha =0.4,edgecolor ="black")
plt.title(" Top 10 startup company's in india ")
plt.xlabel("City")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.grid(True)

```

```python
import matplotlib.pyplot as plt

Demo = data["vertical"].value_counts().head(10)

color = ["red","black","pink","orange","green",
         "brown","yellow","purple","skyblue","blue"]

plt.figure(figsize=(10,5))
plt.barh(Demo.index[::-1], Demo.values[::-1], color=color,alpha=0.6,edgecolor="black")

plt.title("Top 10 Startup Sectors",color="red")
plt.xlabel("Number of Startups",color="green")
plt.ylabel("Sector",color="blue")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

demo = data["round"].value_counts().head(7)
plt.figure(figsize=(10,5))

colors = ["skyblue", "lightgreen", "orange", "pink", "red", "brown", "gray"]


plt.pie(demo.values,labels=demo.index, colors=colors,autopct="%1.1f%%")
plt.title("Rounds Distribution percentage wise ")
plt.axis("equal")
plt.show()
```

```python
#how are startup funding amounts distrubution---histogram( take the values in amount column)
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(data["amount"],bins=10,edgecolor="black")
plt.title("distrbution startup funding ")

```

```python
#the relationship between funding year and funding amount---scatterr plot (column--year)

import pandas as pd
import matplotlib.pyplot as plt


plt.scatter(data["year"],data["amount"],color="skyblue", )


plt.title("Funding Amount By Year")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.grid(True, linestyle="--")
plt.show()

```

```python
#which funding round occurs most frequently
import seaborn as sns
demo = data["round"].value_counts().head(10).index

sns.countplot(data, y="round",order = demo, palette = "viridis")
plt.title("Funding Round Distribution")
plt.xlabel("Count")
plt.ylabel("Funding Round")
plt.show()
```
